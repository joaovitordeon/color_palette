```python
import matplotlib.pyplot as plt
import matplotlib
```

 **see all palettes ===>** https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

```python
# Nome da palette vem aqui (ex: hsv)
cmap = matplotlib.cm.hsv
```

```python
mini=0
maxi=16
alpha=0.48

norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
```

```python
colors = [cmap(norm(value),alpha=alpha) for value in range(mini,maxi)]
```

```python
fig, ax = plt.subplots(figsize=(16,7));
ax.imshow([colors]);

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.tight_layout()
plt.yticks([])
plt.xticks(range(mini,maxi),colors , rotation='vertical' , fontsize=12, weight='bold')
plt.draw()
```
